```python
# List of teams and years to iterate through for file names
team_search_queries = {'giants': ["2012","2013"],
                      'rangers': ["2012","2013"],
                      'islanders': ["2012","2013"],
                      'devils': ["2012","2013"],
                      'knicks': ["2012","2013"],
                      'mets': ["2012","2013"],
                      'yankees': ["2012","2013"]}

months = list(range(1,13))
```

```python
working_directory = os.getcwd()
working_directory
```

```python
for team, years in team_search_queries.items():

        # Get the current file working directory
        working_directory = os.getcwd()

        for year in years:
        ### Check to see if the output csv file exists in the working directory already

            for month in months:
                # include a try/pass clause so that it keeps running through each file
                try:

                    csv_file_name = "\\Data\\pysent_scores\\{}_{}_{}_pysent.csv".format(month,team, year)
                    file_path = working_directory + csv_file_name

                    # does_it_exist is True if there is a file path that exists
                    does_it_exist = exists(file_path)

                    # If the file already exists then move to the next month
                    if does_it_exist == True:
                        print("The file for team: {}, year: {} already exists".format(team, year))
                        continue

                    # If the file does not already exist - then run the code
                    else:
                        # Print out to let me know progess
                        print("Currently analysing: {} {} for {}".format(month, year, team))

                        # Access the correct year/team file
                        file_dir = r"C:\Users\clark\OneDrive - University of Warwick\Diss\1. Twitter API\Data\Data - Scraped Tweets\FINAL Tweets\Reformatted Monthly Team Tweets"
                        file = "\{}_{}_{}_tweets_FINAL.csv".format(month,team,year)

                        file_path = file_dir + file

                        unclean_data = pd.read_csv(file_path)
                        #display(unclean_data.head())


                        # Create the pysent score files for each team/year
                        get_pysent_scores(unclean_data, file_name = "{}_{}_{}_pysent".format(month,team, year))

                # Get the exception to print out the error and for which team/time then pass to next month
                except Exception as e:
                    print(e)
                    print("There was an error with :{} {} {}".format(month,team,year))
                    pass
```
