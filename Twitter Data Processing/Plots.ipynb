### Plot 1: Pysent prediction errors

```python
# Create figure and subplots
fig, ax = plt.subplots(figsize = (8.27, 3))

# Select the data
X = df.index 
Y = df["norm_sum_pysent"]
plt.plot(X,Y, linewidth = 1, color = 'blue')
# Add a black line at zero
plt.hlines(0, xmin = 0, xmax = 366, linewidth = 0.5, color = 'black')

month_starts = ['2012-01-01','2012-02-01','2012-03-01','2012-04-01','2012-05-01','2012-06-01','2012-07-01','2012-08-01','2012-09-01','2012-10-01','2012-11-01','2012-12-01']
month_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'] 

# Change the x-axis ticks
ax.set_xticks(month_starts)
ax.set_xticklabels(month_names)
ax.text(6, 2.5, "A", fontsize = 20)

# Set y-axis label
plt.ylabel("Sentiment prediction errors", fontsize = 8)
plt.xlabel("Date", fontsize = 10)


# Set the margins
plt.margins(x = 0, y = 0.05)

fig.tight_layout()

# plt.show()

save_loc = saving_directory + "\psyent_pe"
#plt.savefig(save_loc)
```

## Plotting relationship between sentiment scores

```python
sns.set(rc={"figure.figsize":(2, 2),
            'xtick.major.pad':"-3", 'ytick.major.pad':"-3",
           'xtick.minor.pad':"-20"}) 
sns.set_style("white")
sns.color_palette()
plt.tight_layout()
```

### Plot 5: Pysent vs VADER

```python
# Pysent vs VADER 
p = sns.lmplot(x = "norm_sum_vader", y = "norm_sum_pysent", data = df ,height  = 2.5,
               aspect = 1, 
               line_kws = {'color' :'black', 'lw': 1} ,
               scatter_kws = {'color' : 'gray', 'alpha': 1, 's': 1})

plt.xlabel("Auxiliary sentiment p.e.", fontsize = 10, labelpad = 2)
plt.ylabel("Primary sentiment p.e.", fontsize = 10, labelpad = -2)

ticks = [-2,0,2]
plt.xticks(fontsize = 8, ticks = ticks)
plt.yticks(fontsize = 8, ticks = ticks)
#plt.set_xticks(ticks)

plt.text(-3, 2.5, "A", fontsize = 15)
plt.margins(x = 0, y = 0.05)

save_loc = saving_directory + "\\Psyent vs Vader"
plt.savefig(save_loc, bbox_inches = "tight")
```

### Plot 6: Surprise cross-reference

```python
# Roberta vs Distilbert 
p = sns.lmplot(x = 'norm_sum_distilbert', y = 'norm_sum_roberta', data = df ,height  = 2.5,
               aspect = 1, 
               line_kws = {'color' :'black', 'lw': 1} ,
               scatter_kws = {'color' : 'gray', 'alpha': 1, 's': 1},
               robust=True)

plt.xlabel("Auxiliary surprise score", fontsize = 10, labelpad = 2)
plt.ylabel("Primary surprise score", fontsize = 10, labelpad = -2)

ticks = [-2,0,2,4]
plt.xticks(fontsize = 8, ticks = ticks)
plt.yticks(fontsize = 8, ticks = ticks)
#plt.set_xticks(ticks)

plt.text(4, 2.5, "B", fontsize = 15)
plt.margins(x = 0, y = 0.05)

save_loc = saving_directory + "\\Surprise cross references"
plt.savefig(save_loc, bbox_inches = "tight")
```
